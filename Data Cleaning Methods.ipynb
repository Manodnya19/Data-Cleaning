```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('C:/Users/manodnyagaikwad/Downloads/Building_Permits.csv')
```

```python
data.head()
```

```python
#calculate the total number of cells in each column which are empty
missing_data = data.isnull().sum()
missing_data
```

```python
#calculate the percentage of missing data
total = np.product(data.shape)
missing_sum = missing_data.sum()

(missing_sum/total) * 100
```

# Dropping columns with NA

```python
#drop all rows with NA
data.dropna()
```

```python
#drop the columns with NAN
columns_withoutNA = data.dropna(axis=1)
columns_withoutNA.head()
```

```python
print("Columns in original dataset: %d \n" % data.shape[1])
print("Columns with na's dropped: %d" % columns_withoutNA.shape[1])
```

# Filling missing data

```python
#Here we will take a subset of the dataset to fill the missing data
subset_data = data.loc[:,'Issued Date':'TIDF Compliance']
subset_data.head()
```

```python
#Use bfill method to fill the missing value with the next value and remaining with 0
subset_data.fillna(method = 'bfill', axis=0).fillna(0)
```
